# Effects of Glucosamine Chondroitin with Methylsulfonylmethane (MSM) on Ankle Pain vs Placebo

## Objective
To evaluate the effectiveness of 1500 mg of Glucosamine, 1200 mg Chondroitin with 250 mg Methylsulfonylmethane (MSM) in reducing ankle pain compared to a placebo.

## Hypothesis
1500mg of Glucosamine, 1200mg Chondroitin with 250 mg of Methylsulfonylmethane (MSM) will reduce ankle pain more effectively than a placebo.

##  Null-Hypothesis
1500mg of Glucosamine, 1200mg Chondroitin with 250 mg of Methylsulfonylmethane (MSM) will not reduce ankle pain more effectively than a placebo.

## Methodology

#### Established my average pain level before study on scale of 1-10. Pain level was 4. 

Treatment:

My father placed oral doses of 1500 mg Glucosamine, 1200 mg Chondroitin with 250 mg of MSM in one clear bottle. Placebo sugar pills were placed in another bottle. The pills were identical in size and shape. I was given one bottle without knowing its content and took two pills every morning for 14 days, recording my results in Excel. After 14 days, I switched to the other bottle for the next 14 days and recorded my results in Excel.

Control Variable:
Maintain the same workout routine throughout the entire study to control for the influence of exercise on pain levels and ankle stability.

Day 1: Walking 
Day 2: No workout
Day 3: Soccer (100min)
Day 4: No workout
Day 5: Chest,Shoulders
Day 6: Back, biceps
Day 7: No workout

Repeated untill study was over 

Data Collection:

Recorded daily pain levels on a scale from 1 to 10, with 1 indicating no pain and 10 indicating the worst possible pain. Additionally, recorded whether I experienced instability in my ankle each day (used 1 for "YES" and 0 for "NO" to indicate instability experienced).

### Participants
Two individuals: 
(Myself)
(My Father who gave me pills) 

### Procedure
1. Establish pre trial numbers; average pain level of 4 each day and instability 50% of the time. 
2. Conducted a blind placebo study over a period of four weeks.
3. Alternate between Glucosamine Chondroitin with MSM and a placebo (sugar pill) every two weeks.
4. Record daily pain level and instability on a scale of 1-10. Instability; 1 for "YES" and 0 for "NO" to indicate instability experienced
5. Noted any workouts or activities done each day.

### Data Collection
Record the data in an Excel spreadsheet with columns for date, day, pain level, instability, workout presence, and type of workout.

### Data Analysis
Use Python for data visualization and analysis. Create line plots to track the pain level and instability over time.

## Results
#### Established my average pain level before study on scale of 1-10. Pain level was 4 (moderate pain). Meaning it interfered signitficantly with daily activities. 
Established my average pain level before the study on a scale of 1-10. Pain level was 4 (moderate pain), meaning it significantly interfered with daily activities.
After the placebo study, my average pain experienced was 3.6/10, a 10% drop in pain levels compared to the established average pain level of 4 per day before the study. After supplementing with 1500 mg Glucosamine, 1200 mg Chondroitin with 250 mg Methylsulfonylmethane (MSM), my average pain level was 2.6/10, which is a 35% drop from my recorded average pain of 4/10. Instability, which I experienced 50% of the time prior to the study, increased to 57% of the time during the placebo phase. While supplementing 1500 mg of Glucosamine, 1200 mg Chondroitin with 250 mg of Methylsulfonylmethane (MSM), my instability was reduced to 7% of the time, a substantial decrease of 87% in instability experienced while taking the supplement. 
## Conclusion
These findings suggest that 1500 mg of Glucosamine, 1200 mg Chondroitin with 250 mg of MSM may indeed be effective in reducing ankle pain and instability when compared to a placebo. However, it's important to note that this study had only one participant (self-study), and the results may not be generalizable to a broader population. Further research with a larger and more diverse group of participants is needed to confirm these findings and establish the effectiveness of this supplement in reducing ankle pain. Additionally, factors such as placebo effects and individual variation should be considered in future studies. The analysis did not investigate tendon and cartilage repair, and therefore, the study does not provide insights into these aspects. Nonetheless, this study provides preliminary evidence supporting the potential benefits of Glucosamine Chondroitin with MSM in managing ankle pain and instability.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loaded the data from the Excel file
file_path = 'Glucosamine Chondrotin with MSM on ankle pain VS Placebo.xlsx'
data = pd.read_excel(file_path, skiprows=3)

# Dropped unnecessary columns
data = data.drop(columns=data.columns[6:], axis=1)

# Renamed the columns for better understanding
data.columns = ['Date', 'Days', 'Pain_Level', 'Instability', 'Workout', 'Workout_Type']

# Converted 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the style for plots
sns.set(style="whitegrid")

# Created a line plot for Pain Level over Time
first_14_days = data.head(14)
following_14_days = data.iloc[14:28]

# Plotted Pain Level over Time for the Placebo Study
plt.figure(figsize=(12, 6))
sns.lineplot(x='Days', y='Pain_Level', data=first_14_days, marker='o', color='green', label='Daily Pain Level')
average_pain_level_first_14 = first_14_days['Pain_Level'].mean()
plt.axhline(average_pain_level_first_14, color='purple', linestyle='--', label='Average Pain Level')
plt.title('Pain Level With Placebo')
plt.xlabel('Days')
plt.ylabel('Pain Level')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Plotted Pain Level over Time for the Supplement Study
plt.figure(figsize=(12, 6))
sns.lineplot(x='Days', y='Pain_Level', data=following_14_days, marker='o', color='blue', label='Daily Pain Level')
average_pain_level_following_14 = following_14_days['Pain_Level'].mean()
plt.axhline(average_pain_level_following_14, color='red', linestyle='--', label='Average Pain Level')
plt.title('Pain Level on Glucosamine')
plt.xlabel('Days')
plt.ylabel('Pain Level')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Split the data into two phases for Instability plots
first_14_days_instability = data.head(14)
following_14_days_instability = data.iloc[14:]

# Plotted Instability over Time for the Placebo Study
plt.figure(figsize=(12, 6))
sns.lineplot(x='Days', y='Instability', data=first_14_days_instability, marker='o', color='blue', label='Daily Instability (Placebo)')
average_instability_first_14 = first_14_days_instability['Instability'].mean()
plt.axhline(average_instability_first_14, color='red', linestyle='--', label=f'Average Instability (Placebo): {average_instability_first_14:.2f}')
plt.title('Instability Over Time with Placebo')
plt.xlabel('Days')
plt.ylabel('Instability (1=Yes, 0=No)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Plotted Instability over Time for the Supplement Study
plt.figure(figsize=(12, 6))
sns.lineplot(x='Days', y='Instability', data=following_14_days_instability, marker='o', color='blue', label='Daily Instability (Glucosamine Chondroitin with MSM)')
average_instability_following_14 = following_14_days_instability['Instability'].mean()
plt.axhline(average_instability_following_14, color='red', linestyle='--', label=f'Average Instability (Supplement): {average_instability_following_14:.2f}')
plt.title('Instability Over Time with Glucosamine Chondroitin with MSM')
plt.xlabel('Days')
plt.ylabel('Instability (1=Yes, 0=No)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```
