# 3-Day Player Load Analysis

### Introduction
This analysis aims to identify which players have the highest and lowest workloads over a three-day period of training sessions at a soccer club. We will explore the reasons behind these workloads by examining various performance metrics such as total distance, high-speed distance, maximum speed, heart rate, impacts, and player load.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "player_training_data.xlsx"

df = pd.read_excel(file_path)

df['Player Load '] = pd.to_numeric(df['Player Load '], errors='coerce').fillna(0)
```

```python
player_workload = df.groupby('Player Name').agg({
    'Player Load ': 'sum',
    'Total Distance (Miles)': 'sum',
    'High-Speed Distance (Miles)': 'sum',
    'Max Speed (mph)': 'mean',
    'HR Avg (bpm)': 'mean',
    'Impacts': 'sum'
}).reset_index()

max_workload_player = player_workload.loc[player_workload['Player Load '].idxmax()]
min_workload_player = player_workload.loc[player_workload['Player Load '].idxmin()]

print("Player with Highest Workload:")
print(max_workload_player)

print("\nPlayer with Lowest Workload:")
print(min_workload_player)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(player_workload['Player Name'], player_workload['Player Load '], color='skyblue')
plt.title('Total Workload by Player')
plt.xlabel('Player Name')
plt.ylabel('Player Load ')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(player_workload['Total Distance (Miles)'], player_workload['Player Load '], color='blue')
for i, player in enumerate(player_workload['Player Name']):
    plt.text(player_workload['Total Distance (Miles)'][i], player_workload['Player Load '][i], player, fontsize=9)
plt.title('Total Distance vs. Workload by Player')
plt.xlabel('Total Distance (Miles)')
plt.ylabel('Player Load ')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(player_workload['High-Speed Distance (Miles)'], player_workload['Player Load '], color='red')
for i, player in enumerate(player_workload['Player Name']):
    plt.text(player_workload['High-Speed Distance (Miles)'][i], player_workload['Player Load '][i], player, fontsize=9)
plt.title('High-Speed Distance vs. Workload by Player')
plt.xlabel('High-Speed Distance (Miles)')
plt.ylabel('Player Load ')
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(player_workload['Impacts'], player_workload['Player Load '], color='green')
for i, player in enumerate(player_workload['Player Name']):
    plt.text(player_workload['Impacts'][i], player_workload['Player Load '][i], player, fontsize=9)
plt.title('Impacts vs. Workload by Player')
plt.xlabel('Impacts')
plt.ylabel('Player Load ')
plt.tight_layout()
plt.show()
```

The data collected includes total distance, high-speed distance, maximum speed, heart rate, impacts, and overall player load. Players were analyzed based on the cumulative player load over three days.

3. **Impacts vs. Workload** - This scatter plot visualizes the relationship between the number of impacts a player experienced and their overall workload.

#### Recommendations
